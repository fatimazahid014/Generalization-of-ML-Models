```python
#@title import libraries
import pandas as pd
```

```python
#@title Loading data

df = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%207/HousingData.csv')
df.info()
```

```python
#@title encode categorical variables

_df = pd.get_dummies(df, columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS,RAD,TAX,PTRATIO,B,LSTAT,MEDV
'])
_df.head()
```

```python
#@title separate features and labels DataFrames

features = _df.drop(['car'], axis=1).values
labels = _df[['car']].values
```

```python
#@title Import more libraries
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#@title Instantiate the decision tree
clf = DecisionTreeClassifier()
params = {'max_depth': np.arange(1, 8)}
```

```python
#@title Instantiate the grid search 
clf_cv = GridSearchCV(clf, param_grid=params, cv=5)
```

```python
#@title  train the models:
clf_cv.fit(features, labels)
```

```python
#@title  Print the best parameter
print("Tuned Decision Tree Parameters: {}".format(clf_cv.best_params_))
```

```python
#@title  Print the best R2
print("Best score is {}".format(clf_cv.best_score_))
```

```python
#@title  Assess the best model
model = clf_cv.best_estimator_
model
```
