```python
#@title import libraries

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#@title Loading data

_headers = ['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings']
df = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%207/abalone.csv', names=_headers, index_col=None)
df.info()
```

```python
#@title split the data into 80% for training and 20% for evaluation

training_df, eval_df = train_test_split(df, train_size=0.8, random_state=1)
```

```python
training_df.head()
```

```python
eval_df.head()
```

```python
#@title Importing KFold
from sklearn.model_selection import KFold
```

```python
#@title  Now create an instance of the class
_kf = KFold(n_splits=5)
indices = _kf.split(df)
```

```python
#@title Find out what data type indices has
print(type(indices))
```

```python
#@title Get the first set of indices

train_indices, val_indices = next(indices)
```

```python
#@title Create a training dataset 
train_df = df.drop(val_indices)
```
