```python
#@title Importing modules
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, \
PolynomialFeatures
```

```python
#@title Loading data
_df = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%207/HousingData.csv')
```

```python
#@title Dropping null values
_df.info()
df = _df.dropna()
```

```python
#@title Extract features into a column called X
X = df.drop(['MEDV'], axis=1).values
y = df['MEDV'].values
```

```python
#@title  Split the data into training and evaluation sets
train_X, eval_X, train_y, eval_y = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
#@title Create an instance of a LinearRegression model:
lr_model_1 = LinearRegression()
```

```python
#@title  Fit the model on the training data:
lr_model_1.fit(train_X, train_y)
```

```python
#@title  Use the model to make predictions on the evaluation dataset
lr_model_1_preds = lr_model_1.predict(eval_X)
```

```python
#@title  Print out the R2 score of the model:
print('lr_model_1 R2 Score: {}'.format(lr_model_1.score(eval_X, eval_y)))
```

```python
#@title Print out the Mean Squared Error (MSE) of this model:
print('lr_model_1 MSE: {}'.format(mean_squared_error(eval_y, lr_model_1_preds)))
```

### Engineer Cubic Features

```python
#@title Create a list of tuples to serve as a pipeline:
steps = [
    ('scaler', MinMaxScaler()),
    ('poly', PolynomialFeatures(degree=3)),
    ('lr', LinearRegression())
]
```

```python
#@title  Create an instance of a pipeline
lr_model_2 = Pipeline(steps)
lr_model_2.fit(train_X, train_y)
```

```python
#@title  Print out the R2 score of the model
print('lr_model_2 R2 Score: {}'.format(lr_model_2.score(eval_X, eval_y)))
```

```python
#@title  Use the model to predict on the evaluation data
lr_model_2_preds = lr_model_2.predict(eval_X)
```

```python
#@title  Print the MSE of the second model:
print('lr_model_2 MSE: {}'.format(mean_squared_error(eval_y, lr_model_2_preds)))
```

```python
#@title Inspect the model coefficients
print(lr_model_2[-1].coef_)
```

```python
#@title 
print(len(lr_model_2[-1].coef_))
```

## Engineer polynomial features

```python
#@title Create a steps list with PolynomialFeatures of degree 10:
steps = [
    ('scaler', MinMaxScaler()),
    ('poly', PolynomialFeatures(degree=10)),
    ('lr', LinearRegression())
]
# Create a third model from the preceding steps
lr_model_3 = Pipeline(steps)
```

```python
#@title Fit the third model on the training data
lr_model_3.fit(train_X, train_y)
```

```python
#@title Print out the R2 score of this model
print('lr_model_3 R2 Score: {}'.format(lr_model_3.score(eval_X, eval_y)))
```

```python
#@title  Use lr_model_3 to predict on evaluation data:
lr_model_3_preds = lr_model_3.predict(eval_X)
```

```python
#@title  Print out the MSE for lr_model_3:
print('lr_model_3 MSE: {}'.format(mean_squared_error(eval_y, lr_model_3_preds)))
```

```python
#@title Print out the number of coefficients 
print(len(lr_model_3[-1].coef_))
```

```python
#@title nspect the first 35 coefficients to get a sense of the individual magnitudes
print(lr_model_3[-1].coef_[:35])
```

## Implement Lasso on the same pipeline

```python
#@title Create a list of steps for the pipeline you will create
steps = [
    ('scaler', MinMaxScaler()),
    ('poly', PolynomialFeatures(degree=10)),
    ('lr', Lasso(alpha=0.01))
]
```

```python
#@title  Create an instance of a pipeline:
lasso_model = Pipeline(steps)
```

```python
#@title  Fit the pipeline on the training data
lasso_model.fit(train_X, train_y)
```

```python
#@title Print the R2 score of lasso_model:
print('lasso_model R2 Score: {}'.format(lasso_model.score(eval_X, eval_y)))
```

```python
#@title  Use lasso_model to predict on the evaluation data
lasso_preds = lasso_model.predict(eval_X)
```

```python
#@title Print the MSE of lasso_model
print('lasso_model MSE: {}'.format(mean_squared_error(eval_y, lasso_preds)))
```

```python
#@title  Print out the number of coefficients in lasso_model
print(len(lasso_model[-1].coef_))
```

```python
#@title  Print out the values of the first 35 coefficients:
print(lasso_model[-1].coef_[:35])
```
