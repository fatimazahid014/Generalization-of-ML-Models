```python
#@title import libraries

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#@title Loading data

_headers = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'car']
# read in cars dataset
df = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%207/car.data', names=_headers, index_col=None)
df.info()
```

```python
#@title encode categorical variables

_df = pd.get_dummies(df, columns=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])
_df.head()
```

```python
#@title separate features and labels DataFrames

features = _df.drop(['car'], axis=1).values
labels = _df[['car']].values
```

```python
#@title Importing other modules
from sklearn.linear_model import LogisticRegressionCV
```

```python
#@title Instantiate LogisticRegressionCV
model = LogisticRegressionCV(max_iter=2000, multi_class='auto', cv=5)
```

```python
#@title Fit the model:
model.fit(features, labels.ravel())
```

```python
#@title  Evaluate the training R2
print(model.score(features, labels.ravel()))
```
